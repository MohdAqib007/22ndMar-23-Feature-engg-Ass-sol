```python
#Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. 
#Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. 
#Calculate the Pearson correlation coefficient between these two variables and interpret the result.

#Ans-

'''To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores, you need a dataset that contains pairs of observations for both variables. Let's assume you have such a dataset and denote the time spent studying as variable X and the exam scores as variable Y.

Calculate the means of X and Y. Let's denote them as X̄ and Ȳ, respectively.

Calculate the standard deviations of X and Y. Denote them as σX and σY.

Calculate the covariance between X and Y. Denote it as Cov(X, Y).

Calculate the Pearson correlation coefficient (r) using the formula:

r = Cov(X, Y) / (σX * σY)

Once you have calculated the Pearson correlation coefficient, you can interpret the result based on its value:

If the Pearson correlation coefficient is close to +1, it indicates a strong positive linear relationship. This means that as the amount of time students spend studying increases, their final exam scores tend to increase as well.
If the Pearson correlation coefficient is close to -1, it indicates a strong negative linear relationship. This means that as the amount of time students spend studying increases, their final exam scores tend to decrease.
If the Pearson correlation coefficient is close to 0, it indicates a weak or no linear relationship between the two variables. This means that there is little to no association between the amount of time students spend studying and their final exam scores.
It's important to note that the Pearson correlation coefficient only measures the linear relationship between variables. It does not capture nonlinear relationships or other types of associations that may exist between the variables.

For Example-

Let's assume you have the following dataset with the amount of time students spend studying (X) and their corresponding final exam scores (Y):

X: [4, 6, 8, 5, 7]
Y: [70, 85, 90, 75, 80]

Calculate the means of X and Y:
X̄ = (4 + 6 + 8 + 5 + 7) / 5 = 6
Ȳ = (70 + 85 + 90 + 75 + 80) / 5 = 80

Calculate the standard deviations of X and Y:
To calculate the standard deviation, we need to calculate the variance first. The variance (Var) is the average of the squared differences from the mean.

For X:
Var(X) = [(4 - 6)^2 + (6 - 6)^2 + (8 - 6)^2 + (5 - 6)^2 + (7 - 6)^2] / 5
= (4 + 0 + 4 + 1 + 1) / 5
= 2

σX = sqrt(Var(X)) = sqrt(2) ≈ 1.414

For Y:
Var(Y) = [(70 - 80)^2 + (85 - 80)^2 + (90 - 80)^2 + (75 - 80)^2 + (80 - 80)^2] / 5
= (100 + 25 + 100 + 25 + 0) / 5
= 50

σY = sqrt(Var(Y)) = sqrt(50) ≈ 7.071

Calculate the covariance between X and Y:
Cov(X, Y) = [(4 - 6) * (70 - 80) + (6 - 6) * (85 - 80) + (8 - 6) * (90 - 80) + (5 - 6) * (75 - 80) + (7 - 6) * (80 - 80)] / 5
= (-20 + 0 + 20 - 5 + 0) / 5
= -1

Calculate the Pearson correlation coefficient (r):
r = Cov(X, Y) / (σX * σY)
= -1 / (1.414 * 7.071)
≈ -0.099

The Pearson correlation coefficient between the amount of time students spend studying and their final exam scores is approximately -0.099. 
This indicates a weak negative linear relationship between the two variables. However, since the correlation coefficient is close to 0, the relationship is very weak and may not be statistically significant.
'''
```

```python
#Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
#Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. 
#Calculate the Spearman's rank correlation between these two variables and interpret the result.

#Ans-

'''To calculate Spearman's rank correlation coefficient between the amount of sleep individuals get each night and their overall job satisfaction level, you need a dataset that contains pairs of observations for both variables. Let's assume you have such a dataset and denote the amount of sleep as variable X and the job satisfaction level as variable Y.

Assign ranks to the values of X and Y separately. If there are ties, assign the average rank to those tied observations. Let's denote the ranks of X as RX and the ranks of Y as RY.

Calculate the differences between the ranks of X and Y for each observation. Denote it as D.

Calculate the Spearman's rank correlation coefficient (ρ) using the formula:

ρ = 1 - (6 * Σ(D^2)) / (n * (n^2 - 1))

where n is the number of observations.

Once you have calculated the Spearman's rank correlation coefficient, you can interpret the result based on its value:

If the coefficient is close to +1, it indicates a strong positive monotonic relationship. This means that as the amount of sleep individuals get each night increases, their overall job satisfaction tends to increase as well, or vice versa.
If the coefficient is close to -1, it indicates a strong negative monotonic relationship. This means that as the amount of sleep individuals get each night increases, their overall job satisfaction tends to decrease, or vice versa.
If the coefficient is close to 0, it indicates a weak or no monotonic relationship between the two variables. This means that there is little to no association between the amount of sleep individuals get and their overall job satisfaction, in terms of monotonicity.
Now, let's calculate the Spearman's rank correlation coefficient with an example:

Assume you have the following dataset with the amount of sleep individuals get (X) and their corresponding job satisfaction levels (Y):

X: [7, 5, 8, 6, 6]
Y: [9, 5, 8, 3, 7]

Assign ranks to X and Y separately:
For X: [7, 5, 8, 6, 6]
Ranks: [3, 1, 5, 2.5, 2.5] (Note: We assign average ranks for tied observations)

For Y: [9, 5, 8, 3, 7]
Ranks: [4, 2.5, 3, 1, 2] (Note: We assign average ranks for tied observations)

Calculate the differences between the ranks of X and Y for each observation (D):
D: [3 - 4, 1 - 2.5, 5 - 3, 2.5 - 1, 2.5 - 2]
[-1, -1.5, 2, 1.5, 0.5]

Calculate the Spearman's rank correlation coefficient (ρ):
ρ = 1 - (6 * Σ(D^2)) / (n * (n^2 - 1))
= 1 - (6 * (1^2 + 1.5^2 + 2^2 + 1.5^2 + 0.5^2)) / (5 * (5^2 - 1))
≈ 1 - (6 * (1 + 2.25 + 4 + 2.25 + 0.25)) / (5 * 24)
≈ 1 - (6 * 10.75) / 120
≈ 1

The calculated Spearman's rank correlation coefficient (ρ) is approximately 1. 
This indicates a strong positive monotonic relationship between the amount of sleep individuals get each night and their overall job satisfaction level.

In other words, the data suggests that as the amount of sleep individuals get increases, their overall job satisfaction tends to increase as well. 
This monotonic relationship implies that individuals who get more sleep generally tend to report higher levels of job satisfaction.

It's important to note that Spearman's rank correlation coefficient measures the strength and direction of the monotonic relationship between variables, regardless of the specific numerical values. In this case, the coefficient indicates a strong positive monotonic relationship, but it does not provide information about the magnitude or causality of the relationship.'''
```
